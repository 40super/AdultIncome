```python
# This function will take the epi data and process it by cleaning and choosing the right columns that will be used for analysis
def StataDFProcessing(df , year = 0000):
    '''
     This Functions takes a Dataframe followed by the year to clean the data and apply the year to show what year the data is from.
    '''
    datacolumns = ['region','statefips','age','female','wbhao','educ','faminc','emp','dlfstat','dind03','mocc10','occ18','multjobs','numjobs','metstat','cbsasize','citistat','citizen','famrel','married','veteran','basicwgt']
    df_stg1 = df[datacolumns]
    print(f"Dataframe NaN Summary: \n{df_stg1.isna().sum()}")
    print(f"Dataframe Length Before Cleaning: {len(df_stg1)}")
    print("---------------CLEANING BEGINS----------------")
    print("Dropping NAN")
    df_stg2 = df_stg1.dropna(how='any')
    print("Dropping -1")
    df_stg2 = df_stg2[df_stg2.occ18 != -1]
    print("Resetting Index")
    df_stg2.reset_index(inplace=True)
    print("Adding year_of_data")
    df_stg2['year_of_data'] =  year
    df_stg2 = df_stg2.apply(pd.Series.replace , to_replace = '$75,000 - $99,999 (2003m10-present) / $75,000+ (1989-2003m9)' , value='$75,000 - $99,999')
    df_stg2 = df_stg2.apply(pd.Series.replace , to_replace = '$150,000+ (2003m10-present)' , value='$150,000+')
    df_stg2 = df_stg2.apply(pd.Series.replace , to_replace = '$100,000 - $149,999 (2003m10-present)', value='$100,000 - $149,999')
    df_stg2 = df_stg2.drop('index',axis=1)
    print("---------------CLEANING ENDS----------------")
    print(f"Dataframe Length After Cleaning: {len(df_stg2)}")
    print(f"Dataframe NaN Summary: \n{df_stg2.isna().sum()}")
    return df_stg2
    
```
